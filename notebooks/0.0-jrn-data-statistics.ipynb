```python
import os
import scienceplots
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Adding LATeX to the PATH
os.environ["PATH"] += os.pathsep + '/Library/TeX/texbin'

# Reduce DPI size when using 'ieee' for SciencePlot
plt.rcParams.update({'figure.dpi': '100'})
```

```python
# Load the dataset from the TSV file
filepath = "../data/raw/participants.tsv"
df = pd.read_csv(filepath, sep="\t")
```

```python
# Group distribution
with plt.style.context(['science', 'notebook']):
    plt.figure()
    df['group'].value_counts().plot(kind='bar')
    plt.title('Participant Group Distribution')
    plt.xlabel('Group')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Sex Distribution
with plt.style.context(['science', 'notebook']):
    plt.figure()
    df['sex'].value_counts().plot(kind='bar')
    plt.title('Sex Distribution')
    plt.xlabel('Sex')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Hemisphere Distribution
with plt.style.context(['science', 'notebook']):
    plt.figure()
    df['hemisphere'].value_counts().plot(kind='bar')
    plt.title('Hemisphere Distribution')
    plt.xlabel('Hemisphere Affected')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Lobe Distribution
with plt.style.context(['science', 'notebook']):
    plt.figure()
    df['lobe'].value_counts().plot(kind='bar')
    plt.title('Lobe Distribution')
    plt.xlabel('Lobe Affected')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Age distribution at epilepsy onset
with plt.style.context(['science', 'notebook']):
    plt.figure()
    df['age_epilepsyonset'].value_counts().sort_index().plot(kind='bar')
    plt.title('Age at Epilepsy Onset Distribution')
    plt.xlabel('Age Category')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
with plt.style.context(['science', 'notebook']):
    plt.figure()
    df['mri_diagnosis'].value_counts().plot(kind='bar')
    plt.title('MRI Diagnosis Distribution')
    plt.xlabel('MRI Diagnosis')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Postoperative outcome distribution
outcome_columns = ['1year_outcome', 'latest_outcome']
outcome_data = df[outcome_columns].apply(pd.Series.value_counts).T.fillna(0)
with plt.style.context(['science', 'notebook']):
    plt.figure()
    outcome_data.plot(kind='bar', stacked=True)
    plt.title('Postoperative Outcome Distribution (Engel Classification)')
    plt.xlabel('Outcome Type')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Correlation heatmap of numerical data
correlation_columns = ['age_epilepsyonset', 'age_scan', 'age_op', 'time_follow-up']
correlation_matrix = df[correlation_columns].corr()
with plt.style.context(['science', 'notebook']):
    plt.figure()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.show()
```
